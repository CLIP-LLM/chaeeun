```python
import torch
import clip
from PIL import Image
from transformers import CLIPProcessor, CLIPModel

# Hugging Face에서 fashion-clip 모델 다운로드 및 로드
model = CLIPModel.from_pretrained("patrickjohncyh/fashion-clip")
processor = CLIPProcessor.from_pretrained("patrickjohncyh/fashion-clip")
```

```python
# 예시 이미지 로드
image = Image.open('/content/drive/MyDrive/sample1.png')

# 예시 텍스트
texts = ["Autumn dress code", "brown jacket", "white balman shoes", "black bag"]

# 이미지를 전처리하고 텍스트와 함께 모델에 입력
inputs = processor(text=texts, images=image, return_tensors="pt", padding=True)

# 모델에서 로짓(logits)을 계산
outputs = model(**inputs)
logits_per_image = outputs.logits_per_image  # 이미지-텍스트 유사도

# 가장 유사한 텍스트 찾기
probs = logits_per_image.softmax(dim=1)  # softmax를 적용하여 확률로 변환
print("Text probabilities:", probs)

# 가장 높은 확률을 가진 텍스트의 인덱스 찾기
max_prob_index = torch.argmax(probs, dim=1).item()

# 해당 인덱스에 대응하는 텍스트 출력
print(f"The most probable text: {texts[max_prob_index]}")
```
