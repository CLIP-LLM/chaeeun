```python
import torch
import clip

# CLIP 모델 로드, 패치 32와 16 중에 선택
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-B/32", device=device)
```

```python
# 청크로 나눈 텍스트들의 리스트
texts = [split.page_content for split in splits]

# clip.tokenize로 텍스트 토큰화
text_inputs = clip.tokenize(texts,truncate=True).to(device)

# 텍스트 임베딩 생성
with torch.no_grad():
    text_embedding = model.encode_text(text_inputs)
```

```python
import chromadb

# Chroma 인스턴스 및 컬렉션 생성
client = chromadb.Client()
collection = client.create_collection("my_collection")

# 메타데이터 생성 (예: 각 문서와 매칭되는 정보)
metadata = [{"id": str(i), "text": split.page_content} for i, split in enumerate(splits)]

# Chroma 컬렉션에 임베딩과 메타데이터 추가
collection.add(
    embeddings=text_embedding.tolist(),  # 임베딩 넘겨줌
    metadatas=metadata,                   # 메타데이터 넘겨줌
    ids=[str(i) for i in range(len(text_embedding))]  # 각 임베딩에 고유 ID 부여
)
```

###clip 모델 이미지 분류 테스트

```python
from PIL import Image
import torch

from transformers import CLIPProcessor, CLIPModel

model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

## 이미지 경로 지정
image_path = '/content/drive/MyDrive/산학협력 프로젝트/Bernese-Dağ-Köpeği-Bernese-Mountain-Dog-Özellikleri.jpg'

image = Image.open(image_path).convert('RGB')

## 텍스트 쿼리 지정
text_query = ["a photo of a retriever", "a photo of a poodle", "a photo of a Dachshund", "a photo of a bernese mountain dog"]

inputs = processor(text=text_query, images=image, return_tensors="pt", padding=True)

outputs = model(**inputs)
logits_per_image = outputs.logits_per_image
probs = logits_per_image.softmax(dim=1)

pred_idx = torch.argmax(probs[0]).item()
pred_text = text_query[pred_idx]

print(f'pred : {pred_text}')

# 이미지를 인식하고 텍스트 쿼리 중에서 구별할 수 있음.
```
